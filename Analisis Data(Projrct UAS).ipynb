```python
import pandas as pd

df = pd.read_csv("pokemon_data.csv")
df.head()

```

```python
stats_split = df['stats'].str.extract(
    r"hp=(\\d+), attack=(\\d+), defense=(\\d+), special-attack=(\\d+), special-defense=(\\d+), speed=(\\d+)"
)
stats_split.columns = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']
stats_split = stats_split.astype(float)
df_clean = pd.concat([df, stats_split], axis=1)
```

```python
numerical_data = df_clean[['base_experience', 'height', 'weight', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
print(numerical_data.describe())
```
